## Problem Definition

<p style="text-align: justify; font-size: 20px;">
The problem selected was to demonstrate the relation between internet access/connection quality and the education of people with disabilities. Online education tools have facilitated the process for people with disabilities. As an example, People with disabilities in intellectual capacities may be at a societal disadvantage due to having fewer educational opportunities in traditional education[1]. A study in Saudi Arabia has shown that most parents viewed online education as beneficial for their children with learning disabilities[3]. Online education is helpful but not limited to intellectual disabilities. According to the ability of visually impaired people with restricted eyesight, internet connectivity is used to access files in the form of audio, visual, or text with the aid of several software that can convert text into audio[2]. In this case, we will also examine the relationship between internet access and the education of visually impaired citizens in Istanbul by analyzing and plotting from the public data. </p>


## DATA

<p style="text-align: justify; font-size: 20px;">
Following DATA were gathered from the Istanbul Büyükşehir Belediyesi (IBB) website open data portal (data.ibb.gov.tr), which continued information regarding the satisfaction from the internet, the number of visually impaired citizens in each district and their genders, and the education status of the visually impaired citizens in each district. To read and visualize the data available as a .xlsx file, python libraries such as numpy, math, mathplotlib, and pandas were used in addition to core Python functionalities. Throughout the data being used in this proposition, preprocessing step was relatively easier since it was in a clean form and did not contain any empty fields regarding values. The only adjustment made was related to basic excel regulations such as renaming the column names if they continued special characters that were creating issues during extraction with Python. 
</p>




```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = '9-vdym-ilce-baznda-hanelerin-ev-internetine-dair-memnuniyet-dereceleri.xlsx'  # Update with the correct file path
data = pd.read_excel(file_path)

# Strip leading and trailing spaces from the column names
data.columns = data.columns.str.strip()

# Verify column names after stripping spaces
print(data.columns)

# Group the data by 'İlçe' and calculate the sum of satisfaction columns for each district
sum_satisfaction = data.groupby('İlçe')[['Hız_Memnunum', 'Hız_Ne Memnunum Ne Memnun Değilim', 'Hız_Memnun Değilim']].sum().reset_index()

# Create the bar chart
plt.figure(figsize=(14, 8))
bar_width = 0.25  # Width of the bars

# Set positions of bars on X axis
r1 = range(len(sum_satisfaction['İlçe']))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, sum_satisfaction['Hız_Memnunum'], color='green', width=bar_width, label='Hız_Memnunum')
plt.bar(r2, sum_satisfaction['Hız_Ne Memnunum Ne Memnun Değilim'], color='blue', width=bar_width, label='Hız_Ne Memnunum Ne Memnun Değilim')
plt.bar(r3, sum_satisfaction['Hız_Memnun Değilim'], color='lightcoral', width=bar_width, label='Hız_Memnun Değilim')

plt.xlabel('İlçe')
plt.ylabel('Total Speed Satisfaction')
plt.title('Total Speed Satisfaction by Istanbul Districts (İlçe)')
plt.xticks([r + bar_width for r in range(len(sum_satisfaction['İlçe']))], sum_satisfaction['İlçe'], rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

The graph takes into account 2 main columns from the Excel sheet, "İlçe" and "Hız_Memnuniyet" which hold the names of the districts and the satisfaction point regarding households in terms of speed satisfaction.

```python
import pandas as pd
import matplotlib.pyplot as plt
#PieChart for each district

# Load the Excel file
file_path = 'cinsiyet-durumuna-gore-ilce-bazl-gorme-engelli-vatanda-says.xlsx'  # Update with the correct file path
data = pd.read_excel(file_path)

# Strip leading and trailing spaces from the column names
data.columns = data.columns.str.strip()

# Verify column names after stripping spaces
print(data.columns)

# Iterate through each district (İlçe) and create a pie chart
districts = data['İlçe'].unique()
for district in districts:
    district_data = data[data['İlçe'] == district]
    kadın = district_data['Görme Engelli Kadın Vatandaş Sayısı'].sum()
    erkek = district_data['Görme Engelli Erkek Vatandaş Sayısı'].sum()
    
    # Create a pie chart
    labels = ['Kadın', 'Erkek']
    sizes = [kadın, erkek]
    colors = ['lightcoral', 'lightskyblue']
    explode = (0.1, 0)  # explode the 1st slice (Kadın)

    plt.figure(figsize=(5, 5))
    patches, texts, autotexts = plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
                                        shadow=True, startangle=140)
    
    # Add count numbers below the percentages
    for i, (text, autotext) in enumerate(zip(texts, autotexts)):
        if i == 0:
            autotext.set_text(f'{autotext.get_text()} Kadın\n({kadın})')
        elif i == 1:
            autotext.set_text(f'{autotext.get_text()} Erkek\n({erkek})')
    
    plt.title(f'Görme Engelli Vatandaş Sayısı in {district}')
    plt.show()
```

## Methodology: Training Methods & Metric Evaluations

<p style="text-align: justify; font-size: 20px;">
In this section, we will be evaluating how we trained our data and our results based on the metrics. Naive Bayes and Random Forest Regression were used to train our models for different stats such as finding the ratio of visually impaired citizens in Istanbul and the literate ratio of these citizens based on districts.
</p>


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error

# Load the Excel file
file_path = 'cinsiyet-durumuna-gore-ilce-bazl-gorme-engelli-vatanda-says.xlsx'
data = pd.read_excel(file_path)

# Strip leading and trailing spaces from the column names
data.columns = data.columns.str.strip()

# Prepare the data
data['total'] = data['Görme Engelli Kadın Vatandaş Sayısı'] + data['Görme Engelli Erkek Vatandaş Sayısı']
X = data['İlçe']
y = data['total']

# Encoding categorical variables
X_encoded = pd.get_dummies(X)

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Define hyperparameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize Random Forest regressor
rf_model = RandomForestRegressor(random_state=42)

# Perform grid search to find optimal hyperparameters
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_

# Initialize Random Forest model with best parameters
optimized_rf_model = RandomForestRegressor(**best_params, random_state=42)

# Fit the model on the training data
optimized_rf_model.fit(X_train, y_train)

# Making predictions
y_pred_rf = optimized_rf_model.predict(X_test)

# Calculating Mean Squared Error (MSE) as a measure of model accuracy
mse_rf = mean_squared_error(y_test, y_pred_rf)

# Calculate the error margin as the standard deviation of the residuals
error_margin_rf = np.sqrt(mse_rf)

# Calculate the range of the target variable
range_y = np.max(y_test) - np.min(y_test)

# Calculate the MSE of a baseline model (predicting the mean of y)
baseline_prediction = np.mean(y_train)
mse_baseline = np.mean((baseline_prediction - y_test) ** 2)

# Prepare the data for the pie chart
districts = data['İlçe'].unique()
sizes = []
labels = []
colors = []
explode = []

# Define colors for the slices
color_palette = plt.get_cmap('tab20')

for i, district in enumerate(districts):
    district_data = data[data['İlçe'] == district]
    kadın = district_data['Görme Engelli Kadın Vatandaş Sayısı'].sum()
    erkek = district_data['Görme Engelli Erkek Vatandaş Sayısı'].sum()
    total = kadın + erkek
    
    sizes.append(total)
    labels.append(district)
    colors.append(color_palette(i % 20))  # Loop through color palette
    explode.append(0.1)  # explode all slices for visibility

# Set font size for the labels
plt.rcParams['font.size'] = 6

# Create the pie chart
plt.figure(figsize=(14, 14))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140, textprops={'fontsize': 8, 'va': 'center', 'ha': 'center'})
plt.title('İstanbuldaki İlçelere Bağlı Görme Engelli Vatandaş Oranı', fontdict={'fontsize': 16, 'fontweight': 'bold'})

# Add the text box with the Random Forest results
# Rastgele Orman sonuçlarıyla metin kutusunu ekleyin
textstr = '\n'.join((
    f'Mean Squared Error (Random Forest): {mse_rf:.2f}',
    f'Hata Marjı (Random Forest): {error_margin_rf:.2f}',
    f'Hedef Değişkenin Aralığı (y_test): {range_y:.2f}',
    f'Temel Modelin MSE\'si: {mse_baseline:.2f}',
    f'Random Forest Modeli ile Temel Model Karşılaştırması:',
    f'MSE Karşılaştırması: {"Gelişim" if mse_rf < mse_baseline else "Sınırlı Gelişim"}',
    f'Hata Marjı vs. Aralık: {"Doğru Tahminler" if error_margin_rf < range_y else "İyileştirme Potansiyeli"}',
    f'Random Forest modeli, İstanbul\'un çeşitli ilçelerinden gelen verilerle eğitilmiş bir dizi karar ağacı kullanarak görme engelli vatandaşların toplam sayısını doğru bir şekilde tahmin eder.', 
    f'Birden çok ağacın tahminlerini bir araya getirerek, sağlam ve güvenilir tahminler sağlar.'
))

plt.gcf().text(0.5, 0.01, textstr, ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

plt.show()


```

<p style="text-align: justify; font-size: 20px;">
In the pie chart visualization above, the Random Forest model is used to obtain multiple decision trees in training and outputting an enhanced accuracy. DATA was split into parts and a hyperparameter is used in the first step. Throughout the code, to find the optimal parameters, GridSearch is used in our Random Forest Tree. Mean Square Error, Error Margins, and variable Range were calculated with the best parameters found. In the last step, a pie chart graph with our optimal Random Forest Tree is displayed for the visual as well as the MSE, Error Margins, and other aspects. In terms of evaluation metrics, potential improvements, and differences were also displayed by making comparisons with the base model, where it stated that there were improvements in the MSE, a lower MSE than the base model which is another indicator that the model is improved in terms of performance, and improved guesses in terms of Error Margins signifying its reliability in guessing the total number of visually impaired citizens.
</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix

# Load the Excel file
file_path = 'eitim-durumuna-gore-ilce-bazl-gorme-engelli-vatanda-says.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1', skiprows=2)

# Convert "Görme Engelli Vatandaş Sayısı" to numeric
df['Görme Engelli Vatandaş Sayısı'] = pd.to_numeric(df['Görme Engelli Vatandaş Sayısı'], errors='coerce')

# Aggregate the data
okuryazar_degil = df[df['Eğitim Durumu'] == 'Okuryazar Değil'].groupby('İlçe')['Görme Engelli Vatandaş Sayısı'].sum()
others = df[df['Eğitim Durumu'] != 'Okuryazar Değil'].groupby('İlçe')['Görme Engelli Vatandaş Sayısı'].sum()

# Combine into a single DataFrame
combined_df = pd.DataFrame({
    'Okuryazar Değil': okuryazar_degil,
    'Okuryazar': others
}).fillna(0).reset_index()

# Create a target variable based on thresholds
combined_df['Category'] = pd.cut(combined_df['Okuryazar'], bins=[-1, 20, 50, 100, 200, float('inf')], labels=['Çok Düşük', 'Düşük', 'Orta', 'Yüksek', 'Çok Yüksek'])

# Prepare the data for Naive Bayes
X = combined_df[['Okuryazar Değil', 'Okuryazar']]
y = combined_df['Category']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Naive Bayes classifier
nb = GaussianNB()
nb.fit(X_train, y_train)

# Make predictions
y_pred = nb.predict(X_test)

# Evaluate the classifier
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Visualize the results
combined_df['Predicted Category'] = nb.predict(X)

# Plot the data with categories
fig, ax = plt.subplots(figsize=(14, 7))
colors = {'Çok Düşük':'green', 'Düşük':'blue', 'Orta':'yellow', 'Yüksek':'orange', 'Çok Yüksek':'red'}
for category in colors:
    subset = combined_df[combined_df['Predicted Category'] == category]
    ax.bar(subset['İlçe'], subset['Okuryazar'], color=colors[category], label=category)

# Increase font sizes
title_fontsize = 20
label_fontsize = 16
tick_label_fontsize = 12

plt.title('İlçelere Göre Görme Engelli Nüfusunun Okuryazarlık Seviyeleri: Bayesyen Yoğunluk Sınıflandırması', fontsize=title_fontsize)
plt.xlabel('İlçe', fontsize=label_fontsize)
plt.ylabel('Görme Engelli Okuryazar Nüfus', fontsize=label_fontsize)
plt.legend(title='Tahmin Edilen Yoğunluk Kategorisi', fontsize=label_fontsize)
plt.xticks(rotation=45, fontsize=8)  # Decrease the font size of İlçe names
plt.yticks(fontsize=tick_label_fontsize)
plt.tight_layout()
plt.show()
```

<p style="text-align: justify; font-size: 20px;">
Another Method being used with data is the Naive Bayes classifier approach, where it is used to classify the literacy levels of the districts in terms of density. Since each district is independent from one another, Naive Bayes was a suitable model for the visually impaired citizens' data. Similar to the Random Forest, first the data is split into parts and used to train the model, using pythons "GaussianNB". After displaying the metrics, we can see that the overall average of our model is 0.92 (%92), stating the average precision is %92. Another key point that needs to be considered is that for "Yüksek" and "Çok Yüksek" precision is 1 stating our model has %100 precision in terms of these categories but for "Orta" density, the precision is %50, and the "Çok Düşük" density is not displayed due to fewer amount of districts with "Çok Düşük" literate density. The output data for the Naive Bayes results can be seen below.
</p>

## Conclusion: Future Work & Possible Improvements

Although this analysis demonstrates the link between internet connection and education opportunities for disabled citizens, the data we have used is solely based on Istanbul and its district. To enhance the analysis even more and expand it to act as a country-wide analysis, data regarding other provinces of Turkey must be included once they become available as public by their respective municipalities.
